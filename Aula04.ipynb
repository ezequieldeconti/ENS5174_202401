```python
#Criando uma matriz numpy
import numpy as np

#Criando uma matriz de zeros utilizando np.zeros
a= np.zeros((5,5))
print(a)
print('Indexação numpy')
print(a[0,0])

```

```python
#Criando matriz com numeros aleatórios usando np.radom.rand() e mostrando a dimensão da matriz
b = np.random.rand(5,5)
print(b)
print('Dimensão da matriz')
print(b.shape) #shape (dimensão) da matriz
```

```python
#para criar matrizes com numeros minimos e maximos definidos (Borges)
import numpy 
a= numpy.random.uniform(0,1000,50)
a = numpy.around(a,2)
print(a)
```

```python
#Criando matriz com valores aleatorios entre uma faixa estabelecida (Professor)
import numpy

vmin=1 #valor minimo
vmax=100 #valor maximo

c = vmin+(vmax-vmin)*b
c = numpy.around(c,2) #aqui eu uso o numpy.around para definir um numero de casas decimais que eu quiser (x,2)

print(c)
```

```python
#Plotar matriz
import matplotlib.pyplot as plt

plt.plot(c.transpose()) #plota tudo
```

```python
#plotando apenas a primeira linha
x = np.array(range(0,numpy.size(c[0,:])))
print(x)
plt.plot(x,c[0,:]) #usa esse codigo pra plotar a linha desejada, no caso a primeira
plt.title('Grafico com valores inteiros no eixo x')
plt.xticks(x) #definindo os valores no eixo x
```

```python
#Criando um gráfico de barras e exportando

fig, ax = plt.subplots(3,1)
ax[0].bar(x,c[:,0],color='red')
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2],color='green')

#exportar a figura 
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder.replace('\\','/')+'/figuras/'
print(figFolder)
fig.savefig(figFolder+'Figure01.png')
```
